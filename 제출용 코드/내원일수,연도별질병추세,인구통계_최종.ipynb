▼ 내원일수 column 만들기

```python
df_total1 = df[(df['연령구분5세'] == 18) | (df['연령구분5세'] == 17)]  # 모든 성별의 계(17), 남/녀 각각의 소계(18)

# 평균내원일수 column 생성  = '평균내원일수 = 1인당 내원일수'
df_total1['평균내원일수'] = 0                                                 # 평균내원일수(1인당 내원일수) column 생성
df_total1['평균내원일수'] = df_total1['내원일수'] / df_total1['환자수']       # 평균내원일수(1인당 내원일수) 값 입력 (위의 단계를 안하면, error)
```

```python
### 내원일수 높은 top10 질병 추출 code ###

# 여성 1인당 내원일수 top 10 ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ
b = df_total1[df_total1['성별구분'] == 1]         # 여성(소계) 
b0 = b.groupby('항목')['평균내원일수'].sum()/5    # 평균내원일수 모두 더해서, 5개의 연도로 나누기 (1년평균, 1인의 내원일수)
bf = b0.sort_values(ascending=False).head(10)     # 오름차순으로 해서, top10 질병 지정

# 남성 1인당 내원일수 top 10 ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ
a = df_total1[df_total1['성별구분'] == 0]         # 남성(소계)
a0 = a.groupby('항목')['평균내원일수'].sum()/5    # 평균내원일수 모두 더해서, 5개의 연도로 나누기 (1년평균, 1인의 내원일수)
af = a0.sort_values(ascending=False).head(10)     # 오름차순으로 해서, top10 질병 지정

# 전체 1인당 내원일수 top 10ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ
c = df_total1[df_total1['성별구분'] == 2]
c0 = c.groupby('항목')['평균내원일수'].sum()/5
cf = c0.sort_values(ascending=False).head(10) 
```
