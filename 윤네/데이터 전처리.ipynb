# 데이터 전처리

```python
df.rename(columns = {'환자수' : '2017_환자수'}, inplace = True)
df.rename(columns = {'내원일수' : '2017_내원일수'}, inplace = True)
df.rename(columns = {'청구건수' : '2017_청구건수'}, inplace = True)
df.rename(columns = {'요양급여비용총액' : '2017_요양급여비용총액'}, inplace = True)
df.rename(columns = {'보험자부담금' : '2017_보험자부담금'}, inplace = True)

df.rename(columns = {'환자수.1' : '2018_환자수'}, inplace = True)
df.rename(columns = {'환자수.2' : '2019_환자수'}, inplace = True)
df.rename(columns = {'환자수.3' : '2020_환자수'}, inplace = True)
df.rename(columns = {'환자수.4' : '2021_환자수'}, inplace = True)

df.rename(columns = {'내원일수.1' : '2018_내원일수'}, inplace = True)
df.rename(columns = {'내원일수.2' : '2019_내원일수'}, inplace = True)
df.rename(columns = {'내원일수.3' : '2020_내원일수'}, inplace = True)
df.rename(columns = {'내원일수.4' : '2021_내원일수'}, inplace = True)

df.rename(columns = {'청구건수.1' : '2018_청구건수'}, inplace = True)
df.rename(columns = {'청구건수.2' : '2019_청구건수'}, inplace = True)
df.rename(columns = {'청구건수.3' : '2020_청구건수'}, inplace = True)
df.rename(columns = {'청구건수.4' : '2021_청구건수'}, inplace = True)

df.rename(columns = {'요양급여비용총액.1' : '2018_요양급여비용총액'}, inplace = True)
df.rename(columns = {'요양급여비용총액.2' : '2019_요양급여비용총액'}, inplace = True)
df.rename(columns = {'요양급여비용총액.3' : '2020_요양급여비용총액'}, inplace = True)
df.rename(columns = {'요양급여비용총액.4' : '2021_요양급여비용총액'}, inplace = True)

df.rename(columns = {'보험자부담금.1' : '2018_보험자부담금'}, inplace = True)
df.rename(columns = {'보험자부담금.2' : '2019_보험자부담금'}, inplace = True)
df.rename(columns = {'보험자부담금.3' : '2020_보험자부담금'}, inplace = True)
df.rename(columns = {'보험자부담금.4' : '2021_보험자부담금'}, inplace = True)
df
```

# 결측치 해결(성별구분)

```python
df['성별구분'][df['성별구분'] == '여'] = 1 
df['성별구분'][df['성별구분'] == '남'] = 0
```

```python
temp = df.iloc[:,3:]
temp.describe()
temp.replace('-',0,inplace=True, regex=True)
```

# 모두 int로 변환해버리귀

```python
df.iloc[:,3:] = df.iloc[:,3:].replace(',','',regex=True)
df.iloc[:,3:] = df.iloc[:,3:].astype('int')
```
