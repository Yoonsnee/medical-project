```python
import pandas as pd
import os
```

```python
## 해당 경로에 있는 .csv 파일명 리스트 가져오기

path = '/Users/bangjiyun/Desktop/3차 프로젝트/데이터/'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.xls')] ## 파일명 끝이 .csv인 경우'
len(file_list_py)
```

```python
## csv 파일들을 DataFrame으로 불러와서 concat
df = pd.DataFrame()
for i in file_list_py:
    data = pd.read_excel(path + i,skiprows=7)  
    df = pd.concat([df,data])
    
df
```

```python
sum = (df['성별구분']=='계') | (df['연령구분5세']=='소계')
df[sum].reset_index(inplace=True,drop=True)
df[sum].to_csv('/Users/bangjiyun/3rd_project/year/sum.csv', header=True,index=False)
```
