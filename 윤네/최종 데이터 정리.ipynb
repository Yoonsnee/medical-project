#### 컬럼명 바꾸기 : 항목, 성별구분, 연령구분5세, 환자수, 내원일수, 청구건수, 요양급여비용총액, 보험자부담금, (연도, 본인부담금)
#### 쉼표 제거, int로 변환
#### 연령구분5세 라벨링 하기, 대칭되는 값이 무엇인지도 같이 보내기
#### 남은 0, 여는 1, 계는 3으로 int로 변환 

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
raw_data = pd.read_csv('/Users/bangjiyun/Downloads/df_ally_years.csv')
```

```python
raw_data.drop(columns='Unnamed: 0',inplace=True)
```

```python
raw_data
```

```python
raw_data['연령구분5세'][raw_data['연령구분5세'] =='5세미만'] = '005세미만'
raw_data['연령구분5세'][raw_data['연령구분5세'] =='5_9세'] = '05_9세'
raw_data.head(20)
```

```python
# 라벨인코딩
items = [i for i in raw_data['연령구분5세']]
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
raw_data['연령구분5세'] = labels
```

```python
raw_data.head(20)
```

```python
# '계': 17, '소계' : 18, '005세미만': 0, '05_9세': 1, '10_14세': 2, '15_19세': 3, '20_24세': 4, '25_29세' : 5, '30_34세' : 6, '35_39세' : 7, '40_44세' : 8, '45_49세' : 9, '50_54세' : 10, 
# '55_59세' : 11, '60_64세' : 12, '65_69세': 13, '70_74세': 14, '75_79세' : 15, '80세이상' : 16
```

```python
raw_data['성별구분'][raw_data['성별구분'] == 3] = 2
raw_data['성별구분'][raw_data['성별구분'] == 0] = 0
raw_data['성별구분'][raw_data['성별구분'] == 1] = 1
```

```python
raw_data.to_csv('/Users/bangjiyun/3rd_project/year/raw_data.csv')
```
